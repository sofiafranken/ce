```python
# QUESTAO 1

print("\n PERGUNTA 1")
print("Verificando 5 propriedades matemáticas:")

# 1. Propriedade da multiplicação:
print("\n1. Distributiva: a(b + c) = ab + ac")
a = 2
b = 8
c = 4
lado1 = a * (b + c)
lado2 = a * b + a * c
print(f"   {a} × ({b} + {c}) = {lado1}")
print(f"   {a} × {b} + {a} × {c} = {lado2}")
print(f"   Igual? {lado1 == lado2}")

# 2. Propriedade dos expoentes:
print("\n2. Expoentes: a^n × a^m = a^(n+m)")
a = 2
n = 4
m = 7
lado1 = (a**n) * (a**m)
lado2 = a**(n + m)
print(f"   {a}^{n} × {a}^{m} = {lado1}")
print(f"   {a}^({n}+{m}) = {lado2}")
print(f"   Igual? {lado1 == lado2}")

# 3. Propriedade das raízes:
print("\n3. Raízes: raiz(a×b) = raiz(a) × raiz(b)")
a = 4
b = 6
lado1 = (a * b)**0.5
lado2 = a**0.5 * b**0.5
print(f"   raiz({a} × {b}) = {lado1}")
print(f"   raiz({a}) × raiz({b}) = {lado2}")
print(f"   Igual? {abs(lado1 - lado2) < 0.001}")

# 4. Propriedade das desigualdades:
print("\n4. Desigualdades: se a < b, então a + c < b + c")
a = 5
b = 9
c = 2
print(f"   {a} < {b}? {a < b}")
print(f"   {a} + {c} < {b} + {c}? {a + c} < {b + c}? {(a + c) < (b + c)}")

# 5. Propriedade dos logaritmos:
import math
print("\n5. Logaritmos: log(a×b) = log(a) + log(b)")
a = 10
b = 100
lado1 = math.log10(a * b)
lado2 = math.log10(a) + math.log10(b)
print(f"   log({a} × {b}) = {lado1}")
print(f"   log({a}) + log({b}) = {lado2}")
print(f"   Igual? {abs(lado1 - lado2) < 0.001}")
```

```python
# QUESTAO 2

print("PERGUNTA 2")
print("Verificando 3 erros algébricos comuns:")

# Erro 1: (x + a)² NÃO É IGUAL a x² + a²
print("\n1. ERRO: (x + a)² = x² + a²")
x = 3
a = 2
errado = x**2 + a**2
certo = (x + a)**2
print(f"   ERRADO: ({x} + {a})² = {x}² + {a}² = {errado}")
print(f"   CERTO: ({x} + {a})² = {certo}")

# Erro 2: raiz(x² + a²) NÃO É IGUAL a x + a
print("\n2. ERRO: raiz(x² + a²) = x + a")
x = 3
a = 4
errado = x + a
certo = math.sqrt(x**2 + a**2)
print(f"   ERRADO: raiz({x}² + {a}²) = {x} + {a} = {errado}")
print(f"   CERTO: raiz({x}² + {a}²) = {certo:.2f}")

# Erro 3: a/(b + c) NÃO É IGUAL a a/b + a/c
print("\n3. ERRO: a/(b + c) = a/b + a/c")
a = 12
b = 2
c = 4
errado = a/b + a/c
certo = a/(b + c)
print(f"   ERRADO: {a}/({b} + {c}) = {a}/{b} + {a}/{c} = {errado}")
print(f"   CERTO: {a}/({b} + {c}) = {certo}")
```

```python
# QUESTAO 3

print("PERGUNTA 3")
print("Verificando propriedades de vetores:")

import numpy as np

u = np.array([1, 2, 3])
v = np.array([4, 5, 6])
w = np.array([7, 8, 9])
a = 2

print(f"\nVetores:")
print(f"u = {u}")
print(f"v = {v}")
print(f"w = {w}")
print(f"escalar a = {a}")

# 1. Comutativa: u + v = v + u
print(f"\n1. Comutativa: u + v = v + u")
resultado1 = u + v
resultado2 = v + u
print(f"   u + v = {resultado1}")
print(f"   v + u = {resultado2}")
print(f"   São iguais? {np.array_equal(resultado1, resultado2)}")

# 2. Associativa: (u + v) + w = u + (v + w)
print(f"\n2. Associativa: (u + v) + w = u + (v + w)")
resultado1 = (u + v) + w
resultado2 = u + (v + w)
print(f"   (u + v) + w = {resultado1}")
print(f"   u + (v + w) = {resultado2}")
print(f"   São iguais? {np.array_equal(resultado1, resultado2)}")

# 3. Elemento neutro: u + 0 = u
print(f"\n3. Elemento neutro: u + 0 = u")
zero = np.array([0, 0, 0])
resultado = u + zero
print(f"   u + 0 = {resultado}")
print(f"   u = {u}")
print(f"   São iguais? {np.array_equal(resultado, u)}")

# 4. Distributiva: a(u + v) = au + av
print(f"\n4. Distributiva: a(u + v) = au + av")
resultado1 = a * (u + v)
resultado2 = a * u + a * v
print(f"   {a}(u + v) = {resultado1}")
print(f"   {a}u + {a}v = {resultado2}")
print(f"   São iguais? {np.array_equal(resultado1, resultado2)}")
```
