```python
# simulacao:
# A = Energia da rede elétrica (1 = disponível, 0 = falta)
# B = Energia solar armazenada (1 = disponível, 0 = não disponível)
# C = Comando de voz do usuário (1 = ligar a luz, 0 = não ligar)

# valores de entrada simulados
A = int(input("Energia da rede elétrica disponível? (1=Sim, 0=Não): "))
B = int(input("Energia solar armazenada? (1=Sim, 0=Não): "))
C = int(input("Comando de voz para ligar a luz? (1=Sim, 0=Não): "))

# L = (¬A · B · C) + (A · C)
if ((not A and B and C) or (A and C)):
    print("Luz acesa (ligada)")
else:
    print("Luz apagada (desligada)")
```
